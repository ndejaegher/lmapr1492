```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
with MPRester("WOCEDvyZTdTuSHXq") as m:
    structure = m.get_structure_by_material_id("mp-15988")
    
print(" Le matériau à analyser est le Li2CuSb")
```

## Affichage des trois premiers pics

```python
#creation de la figure de diffraction avec la bonne longueur d'onde (1.5406)
xrdCuKa = XRDCalculator(wavelength = 'CuKa1')

#plot du diagramme pour 0 à 45° afin d'y retrouver les trois premiers pics
xrdCuKa.show_plot(structure, two_theta_range=(0,45), annotate_peaks=True, with_labels=True)
```

### Premier pic

```python
xrdPattern = xrdCuKa.get_pattern(structure = structure)
angle1 = xrdPattern.x[0]
intensity1 = xrdPattern.y[0]
hkl1 = xrdPattern.hkls[0][0]['hkl']
multiplicity1 = xrdPattern.hkls[0][0]['multiplicity']

print("L'angle 2*theta est égal à", angle1, "°")
print("L'intensité du pic est de", intensity1)
print("Les indices hkl associés sont", hkl1)
print("La multiplicié est de", multiplicity1)
```

### Deuxième pic

```python
xrdPattern = xrdCuKa.get_pattern(structure = structure)
angle2 = xrdPattern.x[1]
intensity2 = xrdPattern.y[1]
hkl2 = xrdPattern.hkls[1][0]['hkl']
multiplicity2 = xrdPattern.hkls[1][0]['multiplicity']

print("L'angle 2*theta est égal à", angle2, "°")
print("L'intensité du pic est de", intensity2)
print("Les indices hkl associés sont", hkl2)
print("La multiplicié est de", multiplicity2)
```

### Troisième pic 

```python
xrdPattern = xrdCuKa.get_pattern(structure = structure)
angle3 = xrdPattern.x[2]
intensity3 = xrdPattern.y[2]
hkl3 = xrdPattern.hkls[2][0]['hkl']
multiplicity3 = xrdPattern.hkls[2][0]['multiplicity']

print("L'angle 2*theta est égal à", angle3,"°")
print("L'intensité du pic est de", intensity3)
print("Les indices hkl associés sont", hkl3)
print("La multiplicié est de", multiplicity3)
```
